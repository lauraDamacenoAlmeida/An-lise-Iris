## Modelo

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO
import matplotlib.image as mpimg
from sklearn import tree
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import *
from sklearn.metrics import mean_squared_error
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
```

```python

#X = np.array(df[['sepal length','sepal width','petal width','petal length']])
X = df.iloc[:,:4]
Y = np.array(df['class'])

x1,x1_test,y1,y1_test = train_test_split(X,Y,test_size=0.3,random_state=3)

arvore = DecisionTreeClassifier(criterion="gini", max_depth=6)

arvore.fit(x1,y1)
arvore.score(x1_test,y1_test)
```

```python
df.columns[:4]
```

```python
predTree = arvore.predict(x1_test)
predTree
```

```python
print("Recall do modelo: ",metrics.recall_score(y1_test, predTree,average=None)*100)
print('F1 score: ',metrics.f1_score(y1_test, predTree,average=None))
```

```python
print('Report da classificação: ', metrics.classification_report(y1_test,predTree))
```

### Gini e Entropia

Validar o ganho de informação (representa as informações aprendidas sobre os rótulos), onde é considerado a impureza da métrica escolhida.
As impurezas mais comuns nos modelos de classificação são a Entropia e o Gini:

```python
def entropy_criterion(data, labels):
    classes =  np.unique(labels)
    
    s=0
    for c in classes:
        p = np.mean(labels == c)
        s-= p *np.log(p)
    return s
```

```python
entropy_criterion(X,Y)
```

```python
def gini_criterion(data,labels):
    classes =  np.unique(labels)
    
    s=0
    for c in classes:
        p = np.mean(labels == c)
        s+= p * (1-p)
    return s
```

```python
gini_criterion(X,Y)
```

```python
matriz = metrics.confusion_matrix(y1_test,predTree)
plt.rc('figure', figsize = (6,5))
fig, ax = plt.subplots()
classes = df['class'].unique()
im = ax.imshow(matriz,cmap=plt.cm.Blues)
ax.figure.colorbar(im, ax=ax)
ax.set(xticks=np.arange(matriz.shape[1]),
           yticks=np.arange(matriz.shape[0]),
           xticklabels=classes, yticklabels=classes,
           title='Matriz de confusão',
           ylabel='True label',
           xlabel='Predicted label')
thresh = matriz.max().mean()
for i in range(matriz.shape[0]):
    for j in range(matriz.shape[1]):
        ax.text(j, i, format(matriz[i, j]),
                ha="center", va="center",
                color="white" if matriz[i, j] > thresh else "black")
fig.tight_layout()
```
